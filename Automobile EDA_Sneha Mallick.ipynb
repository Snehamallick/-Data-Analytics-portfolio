```python
#11. Are there any anomalies or outliers in the dataset when considering a combination of attributes, and can we identify potential reasons or patterns behind these outliers?
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
potential_outliers = data[(data['price'] < lower_bound) | (data['price'] > upper_bound)]
print("Potential outliers:")
print(potential_outliers[['make', 'price']])
```

```python
#15.  Are there any significant differences in the distribution of car prices across different  consumer preferences, using advanced visualization techniques like heatmaps or choropleth maps?
pivot_table = data.pivot_table(index='body-style', columns=['fuel-type', 'num-of-cylinders'], values='price', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='autumn', annot=True, fmt=".1f", linewidths=.5)
plt.title('Average Car Price Across Different Consumer Preferences')
plt.xlabel('Fuel Type & Aspiration')
plt.ylabel('Car Make')
plt.show()
```
